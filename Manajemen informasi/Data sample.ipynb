```python
import pandas as pd
import numpy as np
import statistics 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('sample.csv')
data
```

```python
sorted_profit = data.sort_values(by='profit', ascending=False)
sorted_profit
```

```python
sorted_profit.describe().T
```

```python
col = ['profit', 'revenue', 'budget', 'runtime', 'vote_count', 'popularity']

for i in col:
    medi = statistics.median(data[i])
    print(f'{i} median: {medi}')
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=sorted_profit.head(10), x='original_title', y='profit')
plt.title('judul film vs profit')
plt.xlabel('judul film')
plt.ylabel('profit')
plt.show()
```
